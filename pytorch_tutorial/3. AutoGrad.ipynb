### <b>AutoGrad Example</b>

```python
x=torch.tensor(6.7,requires_grad=True)
y=torch.tensor(0.0,requires_grad=True)

w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
def binary_cross_entropy(predition,target):
    epsilon=1e-8
    predition=torch.clamp(predition,epsilon,1-epsilon)
    return -((target * torch.log(predition)) + (1-target) * torch.log(1-predition))
```

```python
## Forward Pass 

z=w*x+b

y_pred=torch.sigmoid(z)

loss=binary_cross_entropy(y_pred,y)
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

```python
x.grad.zero_()
```
